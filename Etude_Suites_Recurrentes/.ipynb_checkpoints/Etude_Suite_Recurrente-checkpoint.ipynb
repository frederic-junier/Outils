## Étude d'une suite définie par une relation de récurrence $u_{n+1}=f(u_{n})$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def f(x):
    return 2 * np.sqrt(x + 3)

def graphique_suite_recurrente(xmin = 0, xmax = 0, f = lambda x : x, nmin = 0, 
                                   nmax = 0, unmin = 0, fichier = 'suite-recurrente.pdf',  titre = ''):
    
    def suite_recurrente(nmin, nmax, unmin, f):
        les_un = [unmin]
        for k in range(nmax - nmin):
            les_un.append(f(les_un[-1]))
        return les_un

    def escalier(les_un):
        esc = []
        for k in range(len(les_un) - 1):
            esc.extend([[les_un[k], les_un[k+1]], [les_un[k+1], les_un[k+1]]])
        return np.array(esc)

    fig, ax = plt.subplots()
    fig.set_size_inches((10, 10))    
    les_x = np.linspace(xmin, xmax)
    les_y = np.vectorize(f)(les_x)
    amplitude = les_y.max() - les_y.min()
    ymin, ymax = min(les_y.min() - 0.1 * amplitude, 0 - 0.1 * amplitude)  , les_y.max() + 0.1 * amplitude
    ax.set_ylim(ymin , ymax)    
    ax.set_xlim(xmin, xmax)
    les_n = np.arange(nmin, nmax + 1)
    les_un = suite_recurrente(nmin, nmax, unmin, np.vectorize(f))
    #ax.scatter(les_n, les_un)
    esc = escalier(les_un)
    ax.plot(esc[:,0],esc[:,1],'b-o')
    for k, (x, y) in enumerate(esc):
        if k % 2 == 0:
            ax.annotate(r'$A_{%d}$'%(k//2), xy=(x,y), xycoords='data',
             xytext=(0,+10), textcoords='offset points', fontsize=16)
        else:
            ax.annotate(r'$B_{%d}$'%(k//2), xy=(x,y), xycoords='data',
             xytext=(0,-10), textcoords='offset points', fontsize=16)        
    ax.plot(les_x, les_y, 'r-')
    ax.annotate(r'$y=f(x)$', (les_x[-5], les_y[-5]), xytext=(0,-10), textcoords='offset points', fontsize=16)
    ax.annotate(r'$y=x$', (min(xmax, ymax), min(xmax, ymax)), xytext=(0,-10), textcoords='offset points', fontsize=16)
    ax.annotate(r'$y$', (0, ymax), xytext=(-10,0), textcoords='offset points', fontsize=16)
    ax.annotate(r'$x$', (xmax,0), xytext=(-10,0), textcoords='offset points', fontsize=16)
    ax.plot([xmin,xmax],[xmin,xmax], color='black')
    #positionnemens des axes
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data',0))
    ax.yaxis.set_ticks_position('left')
    ax.spines['left'].set_position(('data',0))
    #titre
    ax.set_title(r'%s'%titre)
    #enregistrement sur disque
    fig.savefig(fichier)
```

```python
graphique_suite_recurrente(xmin = -2, xmax = 10, f = f, nmin = 0, nmax = 10,  unmin = -1, 
                           fichier = 'indice-capacite2-p203.pdf')
```
